```python
df .groupby('city', as_index=False) \
    .agg({'dateAdded': 'count'}) \
    .rename(columns={'dateAdded': 'restaurants'}) \
    .sort_values('restaurants', ascending=False) \
    .iloc[:3]
```

```python
df[df['categories'] == 'Coffee shop'] \
    .groupby('city', as_index=False) \
    .agg({'dateAdded': 'count'}) \
    .rename(columns={'dateAdded': 'restaurants'}) \
    .sort_values('restaurants', ascending=False) \
    .iloc[:5]
```

```python
df['dateAdded'] = pd.to_datetime(df['dateAdded'])
df['dateUpdated'] = pd.to_datetime(df['dateUpdated'])
```

```python
rest_opened_in_october = df[df['dateAdded'].dt.month == 10]
rest_opened_in_october.head(3)
```

```python
df.groupby(df['dateAdded'].dt.month) \
    .agg({'id': 'count'})
```

```python
df['updateDelta'] = df['dateUpdated'] - df['dateAdded']
df.head(3)
```

```python
update_delta_and_max_lat_by_city = df.groupby('city', as_index=False) \
    .agg({'updateDelta': 'mean', 'latitude': 'max'})

berlin_mean = update_delta_and_max_lat_by_city[update_delta_and_max_lat_by_city['city'] == 'Berlin']['updateDelta'][0]
berlin_mean
```
