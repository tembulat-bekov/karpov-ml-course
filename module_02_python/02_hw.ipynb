```python
def circle_square(radius):
    return 3.14 * radius ** 2


circle_square(5)
```

```python
def my_zip(list1, list2):
    result = []
    for i in range(min(len(list1), len(list2))):
        result.append((list1[i], list2[i]))
    return result

my_zip([1, 5, 3, 8, 35], [2, 7, 9])
```

```python
def get_sum(init_sum, interest_rate):
    return init_sum * (100 + interest_rate) / 100


def final_balance(init_sum, interest_rate, years, round_num=2):
    final_balance = init_sum
    for _ in range(years):
        final_balance = get_sum(final_balance, interest_rate)
    return round(final_balance, round_num)
```

```python
print(final_balance(1000, 5, 2))
print(final_balance(700, 7, 2))
```

```python
def math_task(data):
    answer = []
    
    for element in data:
        answer += [element * 3]
     
    for i in range(len(answer)):
        answer[i] = answer[i] % 7
        
    for i in range(len(answer)):
        answer[i] = answer[i] + data[i]
    
    return answer

print(math_task([1, 4, 5, 9]))
```

```python
def sum_as_ints(lst):
    total = 0
    for item in lst:
        try:
            total += int(item)
        except ValueError:
            continue
                
    return total

sum_as_ints(['asdasa', '12', '90', 'wqdqw'])
```

```python
def reversed(array):
    local_arr = array.copy()
    ra = []
    while local_arr:
        ra.append(local_arr.pop())
    return ra
```

```python
arr = [1, 2, 3]

if reversed(reversed(arr)) == arr:
    print("It's okay")
else:
    raise RuntimeError('Function is error')
    
print(arr)
```

```python
def find_substr(substring, string):
    is_find = string.find(substring)
    if is_find != -1:
        return (is_find, is_find + len(substring))
    return -1


print(find_substr('we', 'we can do this'))
print(find_substr('python', 'python is the best language in the world'))
```

```python
def find_element(some_list: list) -> list:
    return some_list[-5::-5]


find_element(['e',6,8,'A','>','^','S','$','R','C',6,'+','#',9,'/',1,'T','!','%','K',7,'-','O','*','<',2,'h',4,'g'])
```

```python
def process_string(string):
    return string[1:].lower().replace('intern', 'junior')

process_string('IIntern reads a lot of books')
```

```python
def check_string(string):
    if string != string.strip():
        return False
    first_word = string.split(' ')[0]
    if not first_word.title() != first_word:
        return False
    return string[-1] == '.'

print(check_string('В этом году будет особенно теплое море.'))
print(check_string('В этом году будет особенно теплое Mоре.'))
print(check_string('В этом году будет особенно теплое море'))
print(check_string(' В этом году будет особенно теплое море'))
```
