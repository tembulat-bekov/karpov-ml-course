# import
We can import some modules like `datetime`

The `datetime` module supplies classes for manipulating dates and times.

```python
# include datetime module into code environment
import datetime

# import module_name
```

```python
# using functions from datetime
christmas_date = datetime.date(year=2025, month=1, day=7)
print(christmas_date)
```

```python
# datetime has many functions working with date and time
some_time = datetime.datetime(2022, 3, 4, 13, 55, 32)

print(f'Year: {some_time.year}')
print(f'Month: {some_time.month}')
print(f'Day: {some_time.day}')
print(f'Hour: {some_time.hour}')
print(f'Minute: {some_time.minute}')
print(f'Second: {some_time.second}')
```

```python
# Timedelta object
delta_1 = datetime.timedelta(days=60)

print(datetime.date(2021, 3, 12) + delta_1)
```

Import with `from` using `collections`

This module implements specialized container datatypes providing alternatives to Pythonâ€™s general purpose built-in containers, dict, list, set, and tuple.

```python
# import collections
# collections.defaultdict

from collections import defaultdict, Counter

# from module_name import object_1, object_2, ...

d = defaultdict(int)

print(d)
print(d['this key does not exist']) # int()
print(d)
```

```python
array = [1, 3, 'a', 'm', 1, 'm', None, 3, 3, ()]

c = Counter(array)
c
```

```python
# Counter working like dictionary
c[1] += 1
print(c)

c['a'] = None
print(c)
```
