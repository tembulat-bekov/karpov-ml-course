# 2. Working with filtering

Filtering using for choose certain values from DataFrame. Condidition in `df[]` contains column (Series) and check all entries value from columns

````python
# df[condition]
df[df['workingday'] == 0]
````

## 2.1 Operator `and`

All conditions must True when we use `and` operator. In Python we write `and`, in Pandas this operator is `&`

````python
df[
    (df['workingday'] == 1) & (df['season'] == 1)
]
````

## 2.2 Operator `or`

````python
df[
    (df['humidity'] < 10) | (df['temp'] > 30.)
]
````

## 2.3 Operator `not`

In Python we using `not` operator, in Pandas operator `not` is  `~`

````python
df[
    ~(df['workingday'] == 0)
]
````

````python
# Operator not, and (~, &)
df[
   (~(df['workingday'] == 0)) & (df['temp'] >= 10) & (df['temp'] < 30)
]
````

# 6. Date and time

In Python we have datetime module. Pandas supports `pd.datetime`.

````python
df['datetime']
````

````python
pd.to_datetime(df['datetime'])
````

````python
df_1 = df.copy()
df_1['datetime'] = pd.to_datetime(df_1['datetime'])
df_1.head()
````

````python
df_1[df_1['datetime'].dt.month == 5]
````

````python
try:
    df[df['datetime'].dt.month == 5]
except:
    print('Not Working')
````

````python
df_1.groupby(
    df_1['datetime'].dt.month
).agg({
    'temp': 'mean',
    'humidity': 'min'
})
````

````python
rents_by_week = df_1.groupby(
    df_1['datetime'].dt.isocalendar().week
).agg({
    'temp': 'count'
})

rents_by_week.sample(10, random_state=42)
````

# 7. Visualization

````python
rents_by_week.plot(
    xlabel='Число недели',
    ylabel='Число поездок',
    title='Количество поездок понедельно',
    grid=True,
    figsize=(16, 9)
)
````

````python
df_1['temp'].hist(bins=100, figsize=(16, 9))
````

````python
ax = rents_by_week.hist(bins=100, figsize=(16, 9))

ax[0, 0].set_title('Гистограмма числа поездок понедельно')
````

````python
ax = df_1.groupby(
    df_1['datetime'].dt.dayofyear
).agg({
    'datetime': 'count'
}).hist(
    bins=100,
    figsize=(16, 9)
)

ax = ax[0, 0]
ax.set_title('Гистограмма количества поездок ежедневно')
ax.set_xlabel('Количество поездок')
ax.set_ylabel('Сколько раз было такое количество поездок')
````
