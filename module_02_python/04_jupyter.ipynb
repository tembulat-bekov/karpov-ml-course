# Jupyter: commands

### Cells

Jupyter notebook have _cell_.

`code` contains Python code that Jupyter executes
`text` contains markdown text
`raw` contains any symbols, Jupyter does not try to execute text in raw

This cell has `text` type. We wrote text here.

`ctrl` + `enter` - start cell
`esc` - exit from cell

### Hotkeys
- `esc` - exit from edit mode
- `b` (as _below_) - create cell below with `code` type
- `a` (as _above_) - create cell above with `code` type
- `y` - make choosen cell `code` type
- `m` (as _markdown_) make `text` type
- `dd` remove choosen cell
- `x` remove choosen cell and save it (like cut)
- `v` paste cell
- `ctrl` + `enter` run cell
- `shift` + `enter` run cell and go below

### Matplotlib

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 9))

x = np.arange(-10, 11, 0.5) # like range(-10, 11), but type is np.array
y = x**2 # each element changed

plt.plot(x, y)
plt.show()
````

````python
y
````

````python
plt.plot(x, y)
plt.grid()
plt.show()
````

````python
plt.plot(x, y)
plt.grid()
plt.title('$y=x^2$')
plt.xlabel('x')
plt.ylabel('y')

plt.savefig('1.pdf')
````

````python
plt.figure(figsize=(16, 9))

plt.scatter(x, y)

plt.plot(x, y, color='r')

plt.grid()
plt.title('$y=x^2$')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
````
