## Loops

### for loop

```python
for a in [9, 16, 25]:
    print('Iteration...')
    print(a ** 0.5 + 1)
```

`for item in iterable()`

```python
for i in range(5):
    print(i)
```

```python
for i in range(1, 7):
    print(i)
```

```python
for i in range(10, 1, -1):
    print(i)
```

```python
for i in range(0, 100, 2):
    if i == 10:
        break
    print(i)
```

## while loop

```python
number = 10_000

while number % 2 == 0:
    print('Iteration...')
    number //= 2
    
print(number)
```

### if condition

```python
number = 4

if number % 2 == 0:
    print('even number')
else:
    print('odd number')
```

```python
accounts_balance = [-1121, 2339, -323, 32554, 2323]

negative_accounts = []

for balance in accounts_balance:
    if balance < 0:
        print(f"{balance} is negative")
        negative_accounts.append(balance)
    print(f"finished processing balance {balance}")
    
print(f"Negative accounts: {negative_accounts}")
```

```python
name = input()

if name == 'Moscow':
    print('The capital of Russia')
elif name == 'London':
    print("The capital of Great Britain")
else:
    print("Unknown")
```

### break, continue

```python
words = ['see', 'this', 'long', 'sentence', 'here', 'no-show', 'no-show']

for i in range(len(words)):
    print(words[i])
    
    if i > 2:
        break
```

```python
for i in range(5):
    for j in range(2):
        print(i, j)
        if j > 2:
            break
```

```python
for feature in ['make sandwitch', 'make coffee', 'watch TV', 'wash plate']:
    if feature == 'make coffee':
        print('Coffe machine is broken, skipping')
        continue
    print(f'Starting section {feature}')
```
